```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('a222-.csv')
import matplotlib.pyplot as plt

# Data
cities = ["Kaohsiung", "New Taipei", "Taipei", "Taoyuan", "Taichung", "Tainan"]
unemployment_rates = [3.40, 3.56, 3.55, 3.50, 3.50, 3.50]
national_avg = 3.36

# Plot
plt.figure(figsize=(8, 5))
bars = plt.barh(cities, unemployment_rates, color=plt.cm.viridis([0.1, 0.3, 0.4, 0.6, 0.7, 0.8]))

# Red line for national average
plt.axvline(national_avg, color='red', linestyle='--', linewidth=2)
plt.text(national_avg + 0.02, -0.5, f'National Avg {national_avg:.2f}%', color='red', fontsize=10)

# Labels and title
plt.xlabel("Unemployment Rate (%)")
plt.title("Taiwan Six-Metropolis Unemployment Rate (2023) – Colonel")
plt.tight_layout()
plt.show()
```
